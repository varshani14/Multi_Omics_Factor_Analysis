```python
import cptac
import pandas as pd
import os

pdac = cptac.Pdac()
print("PDAC loaded!")
```

```python
pdac.list_data_sources()
```

```python
os.makedirs("data", exist_ok=True)

pdac.get_clinical("mssm").to_csv("data/clinical.csv")
print("1. clinical done")

pdac.get_CNV("bcm").to_csv("data/CNV.csv")
print("2. CNV done")

pdac.get_circular_RNA("bcm").to_csv("data/circular_RNA.csv")
print("3. circular_RNA done")

pdac.get_miRNA("bcm").to_csv("data/miRNA.csv")
print("4. miRNA done")

pdac.get_phosphoproteomics("bcm").to_csv("data/phosphoproteomics.csv")
print("5. phosphoproteomics done")

pdac.get_proteomics("bcm").to_csv("data/proteomics.csv")
print("6. proteomics done")

pdac.get_transcriptomics("bcm").to_csv("data/transcriptomics.csv")
print("7. transcriptomics done")

pdac.get_ancestry_prediction("harmonized").to_csv("data/ancestry_prediction.csv")
print("8. ancestry_prediction done")

pdac.get_somatic_mutation("harmonized").to_csv("data/somatic_mutation.csv")
print("9. somatic_mutation done")

pdac.get_followup("mssm").to_csv("data/followup.csv")
print("10. followup done")

pdac.get_medical_history("mssm").to_csv("data/medical_history.csv")
print("11. medical_history done")

pdac.get_cibersort("washu").to_csv("data/cibersort.csv")
print("12. cibersort done")

pdac.get_hla_typing("washu").to_csv("data/hla_typing.csv")
print("13. hla_typing done")

pdac.get_tumor_purity("washu").to_csv("data/tumor_purity.csv")
print("14. tumor_purity done")

pdac.get_xcell("washu").to_csv("data/xcell.csv")
print("15. xcell done")

print("ALL 15 CSV FILES SAVED!")
```

```python
import os
print("Current folder:", os.getcwd())
print("\nFiles in data folder:")
for f in os.listdir("data"):
    size_mb = os.path.getsize(f"data/{f}") / (1024*1024)
    print(f"  {f} - {size_mb:.2f} MB")
```

```python
import shutil
import os

source = r"C:\Users\varsh\data"
destination = r"C:\Users\varsh\Documents\R\cptac_pdac_pca\data"

os.makedirs(destination, exist_ok=True)

for file in os.listdir(source):
    src_path = os.path.join(source, file)
    dst_path = os.path.join(destination, file)
    shutil.move(src_path, dst_path)
    print(f"Moved: {file}")

print("\nAll 15 CSV files moved to:", destination)
```
